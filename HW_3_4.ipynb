```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

df = pd.read_csv("./creditcard.csv")
```

```python
transactions = df["Class"].value_counts(normalize=True)

ypos = np.arange(len(transactions))

plt.figure(figsize=(8,6))
plt.title("супер несбалансированно")
plt.bar(ypos, transactions)
plt.yscale(value="log")
plt.xticks(ypos, ['Not frauds', 'Frauds'])
plt.show()
```

```python
pd.options.display.max_columns = 100
df.head(10)
#значения есть везде
```

```python
X = df.drop("Class", axis=1)
y = pd.Series(df["Class"])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=100, stratify=y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
parameters = [{
    'n_estimators':[10,15],
    'max_features':np.arange(3,5),
    'max_depth':np.arange(4,7)
}]

grid = GridSearchCV(estimator = RandomForestClassifier(random_state=100), param_grid = parameters, scoring = 'roc_auc', cv = 3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
best_model = RandomForestClassifier(max_depth = 6, max_features = 3, n_estimators = 15)
best_model.fit(X_train, y_train)

y_pred = best_model.predict_proba(X_test)
#y_pred
```

```python
y_pred_proba = y_pred[:, 1]

from sklearn.metrics import roc_auc_score

roc_auc_score(y_test, y_pred_proba)
```
